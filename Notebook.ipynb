<h2>Common Join</h2>

```python
#Use inner join to merge sale table and employer together. It will add employers column including region, first name, ect. 
df_emp_id_join = pd.merge(df_sales,df_employers, left_on='emp_id', right_on='emp_id', how='inner')
```

```python
df_emp_id_join
```

```python
#testing to see if it can tell us the sum of week 0 base on region
df_emp_id_join.groupby("region")["w0"].sum().sort_values()
```

```python
#left join. It should only show prod_###
df_prode_code_left_join = pd.merge(df_product,df_sales, left_on='prod_code', right_on='item_code', how='left')
```

```python
df_prode_code_left_join
```

```python
#only prod_###. prod_code column only has prod_###. It will ignored all other kind of item. 
print(df_prode_code_left_join['item_code'].value_counts())
```

```python
df_prode_code_outer_join = pd.merge(df_product,df_sales, left_on='prod_code', right_on='item_code', how='outer')
```

```python
#noticed the null. It will just join everything together even it does not match. If it does not ahve a match, it will add NaN value.
df_prode_code_outer_join
```

```python
df_prode_code_right_join = pd.merge(df_product,df_sales, left_on='prod_code', right_on='item_code', how='right')
```

```python
#The left table will try to join anything that match to the right table. If there is no match, it will add NaN
df_prode_code_right_join
```
